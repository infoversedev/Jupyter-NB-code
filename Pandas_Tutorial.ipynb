```python
# Show example of accumulated sum; cumsum()
df3 = pd.DataFrame({"A":[5, 3, 6, 3], 
                   "B":[11, 2, 17, 3], 
                   "C":[4, 19, 0, 5], 
                   "D":[5, 4, 9, 8]}) 
df3.cumsum(axis=0)
```

```python
# Show example of missing data, Pandas labels as NaN
df3 = pd.DataFrame({"A":[5, 3, 6, 4], 
                   "B":[11, 2, None, 3], 
                   "C":[5, 3, 5, 5], 
                   "D":[5, 4, None, 8]}) 
```

```python
# Example of ignoring missing values in data
df3.cumsum(axis=1, skipna = True)
```
